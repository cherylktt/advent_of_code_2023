```python
lines = ["467..114..",
         "...*......",
         "..35..633.",
         "......#...",
         "617*......",
         ".....+.58.",
         "..592.....",
         "......755.",
         "...$.*....",
         ".664.598.."]

# PART 1
digits = "0123456789"

def split_chars(lines):
    new_lines = []
    for line in lines:
        new_lines.append([x for x in line])
    
    # for y in new_lines:
    #     print(y)
    return new_lines

def get_front(index, line, number=[]):
    if index > 0 and line[index-1] in digits:
        number.insert(0, line[index-1])
        get_front(index-1, line, number)
    return number

def get_back(index, line, number=[]):
    if index < len(line)-1 and line[index+1] in digits:
        number.append(line[index+1])
        get_back(index+1, line, number)
    return number

def get_entire_number(index, line):
    entire_number = ""
    output = get_front(index, line, []) + [line[index]] + get_back(index, line, [])
    for x in output:
        entire_number += x
    return int(entire_number)

def adjacent_symbols(lines):
    part_number = {}
    total_sum = 0
    for i in range(len(lines)):
        for j in range(len(lines[i])):
            if lines[i][j] in digits:
                number = get_entire_number(j, lines[i])
                if i > 0 and j > 0 and lines[i-1][j-1] != "." and lines[i-1][j-1] not in digits:
                    part_number[number] = True
                elif j > 0 and lines[i][j-1] != "." and lines[i][j-1] not in digits:
                    part_number[number] = True
                elif i < len(lines)-1 and j > 0 and lines[i+1][j-1] != "." and lines[i+1][j-1] not in digits:
                    part_number[number] = True
                elif i > 0 and lines[i-1][j] != "." and lines[i-1][j] not in digits:
                    part_number[number] = True
                elif i < len(lines)-1 and lines[i+1][j] != "." and lines[i+1][j] not in digits:
                    part_number[number] = True
                elif i > 0 and j < len(lines[i])-1 and lines[i-1][j+1] != "." and lines[i-1][j+1] not in digits:
                    part_number[number] = True
                elif j < len(lines[i])-1 and lines[i][j+1] != "." and lines[i][j+1] not in digits:
                    part_number[number] = True
                elif i < len(lines)-1 and j < len(lines[i])-1 and lines[i+1][j+1] != "." and lines[i+1][j+1] not in digits:
                    part_number[number] = True
            elif lines[i][j] not in digits:
                if len(part_number) > 0:
                    for k, v in part_number.items():
                        if v == True:
                            total_sum += k
                part_number.clear()
    return total_sum

# PART 2
def get_gear_ratio(lines):
    total_gear_ratio = 0
    for i in range(len(lines)):
        for j in range(len(lines[i])):
            if lines[i][j] == "*":
                adjacent = []
                gear_ratio = 1
                if i > 0 and j > 0 and lines[i-1][j-1] in digits:
                    number = get_entire_number(j-1, lines[i-1])
                    if number not in adjacent:
                        adjacent.append(number)
                if j > 0 and lines[i][j-1] in digits:
                    number = get_entire_number(j-1, lines[i])
                    if number not in adjacent:
                        adjacent.append(number)
                if i < len(lines)-1 and j > 0 and lines[i+1][j-1] in digits:
                    number = get_entire_number(j-1, lines[i+1])
                    if number not in adjacent:
                        adjacent.append(number)
                if i > 0 and lines[i-1][j] in digits:
                    number = get_entire_number(j, lines[i-1])
                    if number not in adjacent:
                        adjacent.append(number)
                if i < len(lines)-1 and lines[i+1][j] in digits:
                    number = get_entire_number(j, lines[i+1])
                    if number not in adjacent:
                        adjacent.append(number)
                if i > 0 and j < len(lines[i])-1 and lines[i-1][j+1] in digits:
                    number = get_entire_number(j+1, lines[i-1])
                    if number not in adjacent:
                        adjacent.append(number)
                if j < len(lines[i])-1 and lines[i][j+1] in digits:
                    number = get_entire_number(j+1, lines[i])
                    if number not in adjacent:
                        adjacent.append(number)
                if i < len(lines)-1 and j < len(lines[i])-1 and lines[i+1][j+1] in digits:
                    number = get_entire_number(j+1, lines[i+1])
                    if number not in adjacent:
                        adjacent.append(number)
                # print(adjacent)
                if len(adjacent) > 1:
                    for x in adjacent:
                        gear_ratio *= x
                    total_gear_ratio += gear_ratio
    return total_gear_ratio

with open ('day3.txt') as file:
    lines = file.read().split("\n")

new_lines = split_chars(lines)
total_sum = adjacent_symbols(new_lines)
total_gear_ratio = get_gear_ratio(new_lines)

print(f"part 1: {total_sum}")
print(f"part 2: {total_gear_ratio}")
```
